```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
x=np.linspace(0,5,11)
x
```

```python
y=x**2
y
```

```python
plt.plot(x,y)
plt.title("Hello")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

Subplot

```python
plt.subplot(2,2,1)
plt.subplot(2,2,2)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,4)
```

Object Oriented way


```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.6,0.8])
axes1.plot(x,y)
axes2=fig.add_axes([0.2,0.5,0.4,0.3])
axes2.plot(x,y)
```
