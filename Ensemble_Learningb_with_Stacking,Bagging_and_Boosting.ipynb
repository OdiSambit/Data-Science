Stacking prediction


```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
import seaborn as sns
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
X=df.drop('species', axis=1)
y=df['species']

le = LabelEncoder()
y_encoded = le.fit_transform(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y_encoded, test_size=0.2, random_state=42,stratify=y_encoded)
```

```python
base_learners = [
    ('dt', DecisionTreeClassifier(random_state=42)),
    ('svm', SVC(probability=True,kernel='rbf', random_state=42)),
    ('lr', LogisticRegression(max_iter=1000))
]
```

```python
meta_learner = LogisticRegression(max_iter=1000)
```

```python
stacking_clf = StackingClassifier(estimators=base_learners, final_estimator=meta_learner,cv=5)
```

```python
stacking_clf.fit(X_train, y_train)
```

```python
y_pred = stacking_clf.predict(X_test)
accuracy =accuracy_score(y_test, y_pred)
accuracy
```

Bagging - Random Forest


```python
#Random Forest
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(
    n_estimators=100, #no of trees
    random_state=42, #let trees grow fully
    max_depth=None)
rf_model.fit(X_train, y_train)
```

```python
y_pred = rf_model.predict(X_test)
accuracy =accuracy_score(y_test, y_pred)
accuracy
```

Till now we are getting better accuracy in Stacking for this data set


Boosting Technique


```python
#Adaboost and Gradient Boosting
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier
```

```python
#XG Boosting
from xgboost import XGBClassifier
```

```python
ada_model = AdaBoostClassifier(n_estimators=100, #no of trees
                               random_state=42)
ada_model.fit(X_train, y_train)
```

```python
y_pred = ada_model.predict(X_test)
accuracy =accuracy_score(y_test, y_pred)
accuracy
```

```python
gradient_model = GradientBoostingClassifier(n_estimators=100, #no of trees
                              learning_rate=0.1, random_state=42)
gradient_model.fit(X_train, y_train)
```

```python
gb_pred = gradient_model.predict(X_test)
accuracy =accuracy_score(y_test, gb_pred)
accuracy
```

```python
Xgb_model = XGBClassifier(n_estimators=100, #no of trees
                              learning_rate=0.1, random_state=42,max_depth=3,use_label_encoder=False,eval_metric='mlogloss')
Xgb_model.fit(X_train, y_train)
```

```python
Xgb_pred = Xgb_model.predict(X_test)
accuracy =accuracy_score(y_test, Xgb_pred)
accuracy
```
