Heat Map


```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
tips_corr = tips[['total_bill','tip','size']]
tips_corr
```

```python
tips_corr.corr()
```

```python
sns.heatmap(tips_corr.corr())
```

```python
sns.heatmap(tips_corr.corr(),annot = True)
```

Cluster Map


```python
sns.clustermap(tips_corr.corr())
```

Pivot table Heat Map

```python
pvtflight = pd.pivot_table(flights,values='passengers',index='month',columns='year')
pvtflight
```

```python
sns.heatmap(pvtflight)
```
